```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 

number={
    '$1000':[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]
}

data=pd.DataFrame(number)
val=data['$1000']
# print(data)
```

```python
# 1 no:

mx=val.max()
mn=val.min()
print(mx-mn)
```

```python
# 2 no:

mean=val.mean()
print(mean,len(val))

res=np.array(val)
print(np.var(res),np.std(res))
res=res-mean
res=res**2
res=np.sum(res)
res=res/len(val)
print(res)
```

```python
# 3 no:

std=res**0.5
print(std)
```

```python
# 4 no:

Q2=val.mean()
print(Q2)
```

```python
# 5 no

Q1=val.quantile(.25)
print(Q1)

Q3=val.quantile(.75)
print(Q3)

IQR=Q3-Q1
print(IQR)
```

```python
# 6 no:

lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR

print(lower_fence,upper_fence)
```

```python
# 7 no

min_outliers=data.loc[(data['$1000']<lower_fence)]
print(min_outliers)

max_outliers=data.loc[data['$1000']>upper_fence]
print(max_outliers)
```
