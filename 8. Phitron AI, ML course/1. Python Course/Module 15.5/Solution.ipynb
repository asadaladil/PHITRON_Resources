```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

student=pd.read_csv("Practice Day.csv")
```

```python
# print(student)
# Q1:

student['Discounted_Price']=student['Price']*0.1
student.to_csv('discounted_products.csv')

# print(student)
```

```python
# Q2:

clean_data=student.drop_duplicates(subset=['EmployeeID']).copy()
print(clean_data)
```

```python
# Q3:

cnt=clean_data.isnull().sum()
print(sum(cnt))

mean=clean_data['Age'].mean()
# print(mean)

clean_data.fillna({'Age':mean},inplace=True)
# print(clean_data)

clean_data.dropna(subset=['Salary'],inplace=True)
print(clean_data)
```

```python
# Q4:

col=['Marks1','Marks2','Marks3']
mean_value={}
print("Mean")
for i in col:
    mx=student[i].mean()
    mean_value[i]=f"{mx:.2f}"
    
for i,j in mean_value.items():
    print(i,j)

max_value={}
print("Max")
for i in col:
    mx=student[i].max()
    max_value[i]=mx

for i,j in max_value.items():
    print(i,j)
    

total_marks=0
for i in col:
    total_marks+=student[i]
    
print(total_marks)

print(f"Standard Deviation={total_marks.std():.2f}")
```
