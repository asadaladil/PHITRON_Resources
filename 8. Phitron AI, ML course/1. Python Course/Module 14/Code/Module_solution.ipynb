```python
import pandas as pd

# Q1:

data = [[3, 108939], [2, 12747], [8, 87709], [6, 91796]]
accounts = pd.DataFrame(data, columns=['account_id', 'income']).astype({'account_id':'Int64', 'income':'Int64'})
print(accounts)
def count_salary_categories(accounts: pd.DataFrame):
    low=len(accounts.loc[accounts['income']<20000])
    avg=len(accounts.loc[(accounts['income']>=20000) & (accounts['income']<=50000)])
    high=len(accounts.loc[accounts['income']>50000])
    
    answer={
        'category':['Low Salary','Average Salary','High Salary'],
        'accounts_count':[low,avg,high]
    }
    df=pd.DataFrame(answer)
    return df

print(count_salary_categories(accounts))
```

```python
# Q2:
data = [[1, 'john@example.com'], [2, 'bob@example.com'], [3, 'john@example.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'int64', 'email':'object'})

print(person)
def delete_duplicate_emails(person: pd.DataFrame):
    person.sort_values(['id'],inplace=True)
    person.drop_duplicates(subset=['email'],inplace=True)
    return person

print(delete_duplicate_emails(person))
```
