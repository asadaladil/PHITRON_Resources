```python
# Q6:

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", 
             "Frank", "Grace", "Henry", "Ivy", "Jack"],
    "Department": ["HR", "Engineering", "Sales", "Engineering", "Sales",
                   "Marketing", "HR", "Engineering", "Sales", "Marketing"],
    "Salary": [70000, 95000, 80000, 110000, 82000,
               68000, 72000, 120000, 81000, 70000]
}

employees=pd.DataFrame(data)
print(employees,end='\n\n')

copy=employees.sort_values(['Salary','Name'],ascending=[1,0])
print(copy)
```

```python
# Q7:

data = {
    "Name": ["Mia", "Liam", "Olivia", "Noah", 
             "Emma", "Oliver", "Ava", "Lucas"],
    "Score": [88, 35, 95, 81, 
              76, 89, 55, 78]
}

students=pd.DataFrame(data)
print(students,end='\n\n')

filtered=students.loc[(students['Score']>60) | (students['Score']<40)]
print(filtered)
```
