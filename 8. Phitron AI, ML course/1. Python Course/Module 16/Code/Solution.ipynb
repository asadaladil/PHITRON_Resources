```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
from matplotlib import pyplot as plt

```

```python

# Q1:

data=pd.read_csv("https://raw.githubusercontent.com/phitronio/Python-for-ML/refs/heads/main/student_IQdata.csv")
print(data)

```

```python
# Q2:

sns.scatterplot(
    data=data,
    x="Study_Hour",
    y="IQ_Score"
)
plt.show()
```

```python
# Q3:

sns.scatterplot(
    data=data,
    x='Shoe_Size',
    y="Study_Hour",
)
plt.show()
```

```python
# Q4:

fig=px.histogram(
    data,
    x='Shoe_Size',
    nbins=13,
    
)
fig.update_traces(       
    marker_line_color='black',   
    marker_line_width=1, 
    # xbins=dict(
    #     size=0.8
    # )
)
fig.show()
```
