```python
import random
import numpy as np
```

```python
# celcius to fahrenheit conversion

celcius=np.array([random.randint(35,50) for i in range(10)])
print(celcius)
fahrenheit=1.8*celcius+32
print(fahrenheit)
```

```python
lap_time=[random.randint(60,90) for i in range(15)]
lap_time=np.array(lap_time)
print(lap_time.shape)
print(lap_time.size)
print(lap_time.dtype)
lap_min=lap_time/60
print(lap_min)
```
