```python
import numpy as np
import pandas as pd
np.random.seed(42)
ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])
DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})
DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

```python
# Question no - 01

ages=np.array(DF['age'])
print(f"Numpy Ages: {ages}")
print(f"Ages Datatype: {ages.dtype}")
print(f"Ages Dimension: {ages.ndim}")
print(f"Ages Shape: {ages.shape}")
print(f"Ages Size: {ages.size}\n")

salaries=np.array(DF['salary'])
print(f"Numpy Salaries: {salaries}")
print(f"Salaries Dimension: {salaries.dtype}")
print(f"Salaries Shape: {salaries.shape}")
print(f"Salaries Size: {salaries.size}")
```

```python
# Question no - 02

print(f"Highest Salary: {np.max(salaries)}")
print(f"Lowest Salary: {np.min(salaries)}")
print(f"Average Salary: {np.mean(salaries)}")
print(f"Average Age: {np.mean(ages)}")
```

```python
# Question no - 03

age_gt30=ages[ages>30]
print(f"Ages > 30: {age_gt30}")
print(f"Count Ages > 30: {age_gt30.size}")
```

```python
# Question no - 04

new_ages=ages+5
print(f"Ages added with 5: {new_ages}")
```

```python
# Question no - 05

print(f"Sum of all salaries: {np.sum(salaries)}")
print(f"Difference between max and min salary: {np.max(salaries)-np.min(salaries)}")
```
